### Gráfico de linha

```python
# Gráfico de  linha

# Dados

x = linspace(0,5,10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0,0,0.8,0.8])

# Cria o plot
axes.plot (x,y,'r')

# Labels e título
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha');

# Exibe gráficos
plt.show()
```

```python
# Gráficos de linha com duas figuras

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Cria os eixos
axes1 = fig.add_axes ((0.1,0.1,0.8,0.8)) # eixos da figura principal
axes2 = fig.add_axes((0.2,0.5,0.4,0.3)) # eixos da figura secundária

# Figura principal
axes1.plot(x,y,'r') # 'r' - significa que a cor do gráfico será vermelha
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Proncipal')

# Figura secundária
axes2.plot(y,x,'g') # 'g' - significa que a cor do gráfico será verde
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Figura Secundária');

# Exibe Gráfico

plt.show()
```

```python
# Gráfico de Linha em Paralelo

# Dados
x = linspace(0,5,10)
y = x **2

# Divide a área de plotagem em dois subplots
fig, axes = plt.subplots(nrows=1,ncols=2)

# Loop pelos eixos para criar cada plot

for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')

# Ajusta o layout ()
fig.tight_layout()

plt.show()
```

```python
# Gráfico de linha com diferentes escalas

# Daddos
x= linspace(0,5,10)
y = x**2

# Cria os subplots
fig, axes, = plt.subplots(1,2, figsize=(10,4))

# Cria o plot1
axes[0].plot(x,x**2,x,exp(x))
axes[0].set_title("Escala Padrão")

# Cria o plot2
axes[1].plot(x,x**2,x,exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logaritmica (y)");

plt.show()
```

```python
# Grid

# Dados
x= linspace(0,5,10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1,2, figsize=(10,3))

# Grid padrão
axes[0].plot(x, x**2,x, x**3, lw=2)
axes[0].grid(True)

# Grid customizado
axes[1].plot(x,x**2,x**3, lw=2)
axes[1].grid(color='b',alpha=0.7, linestyle='dashed',linewidth=0.8)

plt.show()
```

### Histograma

```python
# Histograma

# Dados
n = np.random.randn(100000)

# Cria subplots
fig,axes = plt.subplots(1,2,figsize=(12,4))

# Plot 1
axes[0].hist(n)
axes[0].set_title('Histograma Padrão')
axes[0].set_xlim((min(n),max(n)))

# Plot 2
axes[1].hist(n,cumulative=True,bins=50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)))

# Exibe os gráficos
plt.show()
```

### Gráficos 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
# Dados

alpha = 0.7
phi_ext= 2 * np.pi *0.5

# Função para um mapa de cores 

def ColorMap(phi_m,phi_p):
    return(+ alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext-2*phi_p))

# Mais Dados

phi_m = np.linspace (0,2*np.pi,100)
phi_p = np.linspace(0.2*np.pi,100)
#phi_p = np.linspace(0.2 * np.pi, 2 * np.pi, 100)  # Corrigido para gerar 100 pontos no intervalo correto
X,Y = np.meshgrid(phi_p,phi_m)
Z = ColorMap(X,Y)
#Z = ColorMap(X, Y)  # Não precisa de .T

```

```python
# Cria a figura
fig = plt.figure(figsize=[14, 6])

# Adiciona o subplot1 com projeção 3D
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
p1 = ax1.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# Adiciona o subplot2 com projeção 3D
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
p2 = ax2.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Cria a barra de cores como legenda
fig.colorbar(p2, shrink=0.5)

# Exibe gráfico
plt.show()
```
